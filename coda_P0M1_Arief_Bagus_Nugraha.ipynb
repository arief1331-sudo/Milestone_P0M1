## Import Data Library

```python
import time
import pandas as pd

##scrapping
from selenium import webdriver
from bs4 import BeautifulSoup
```

## Scraping Data

```python
#List Kosong
nama_produk_clean = []
harga_clean = []
pengiriman_clean = []
diskon_clean = []

# Setup driver 
driver = webdriver.Chrome()
url = 'https://www.astronauts.id/search?q=minuman&source=homepage'
driver.get(url)

#agar tidak scroll terus (limiter)
scroll_limit = 5
count = 0

# Ambil ketinggian halaman saat ini (sebelum scroll)
last_height = driver.execute_script("return document.body.scrollHeight")

while True:

    #Memperlambat Scroll

    for i in range(0, last_height, 500):
        driver.execute_script(f"window.scrollTo(0, {i});")
        time.sleep(0.1)

    #Perintah JavaScript untuk scroll sampai ke paling bawah
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # menyesuaikan waktu sleep tergantung kecepatan internet/loading halaman
    time.sleep(5) 

    # Hitung ketinggian halaman baru setelah scroll dan loading
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    count += 1
    print(f"Scroll ke-{count} selesai.")

    #rem
    if count >= scroll_limit:
        print(f"Sudah mencapai batas {scroll_limit} kali scroll. Berhenti paksa.")
        break

    # Cek ketinggian
    if new_height == last_height:
        print("Halaman sudah mencapai akhir.")
        break
    
    # Update last_height untuk berikutnya
    last_height = new_height

print("Sedang mengambil data HTML...")

html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')

boxes = soup.find_all('div',{'class':'MuiBox-root css-1rgpaxo'})

for box in boxes:

    try:
        nama_produk = box.find('span',{'class':'MuiTypography-root MuiTypography-paragraph-tiny css-dndy4p'}) 
        nama_produk_clean.append(nama_produk.get_text())
    except:
        nama_produk_clean.append(None)

    try:
        harga = box.find('span',{'class':'MuiTypography-root MuiTypography-body-smallStrong css-tg7xtj'}) 
        harga_clean.append(harga.get_text())
    except:
        harga_clean.append(None)
    
    try:
        pengiriman = box.find('span',{'class':'MuiTypography-root MuiTypography-caption-tinyBold css-1jvjn92'}) 
        pengiriman_clean.append(pengiriman.get_text())
    except:
        pengiriman_clean.append(None)

    try:
        diskon = box.find('span',{'class':'MuiTypography-root MuiTypography-caption-tinyBold css-148eilu'}) 
        diskon_clean.append(diskon.get_text())
    except:
        diskon_clean.append(None)
    
```

```python
df_data = pd.DataFrame()
```

```python
df_data['Nama Produk']=nama_produk_clean
df_data['harga'] = harga_clean
df_data['pengiriman'] = pengiriman_clean
df_data['diskon'] = diskon_clean


df_data.head(5)
```

```python
df_data.info()
```

```python
df_data.to_csv('milestone_scraping_astro_fix.csv')
```

### Mengdrop column Unnamed: 0

```python
df_astro = df_astro.drop(columns=['Unnamed: 0'])
df_astro
```

### Membersihkan dan mengubah tipe data kolom harga dan diskon

```python
# 1. Membersihkan kolom 'harga'
# Hapus 'Rp'-> hapus titik (.)-> ubah ke integer
df_astro['harga'] = df_astro['harga'].str.replace('Rp', '').str.replace('.', '').astype(int)

# 2. Membersihkan kolom 'diskon'
# Isi yang kosong dengan '0%' -> hapus tanda '%' -> ubah ke float
df_astro['diskon'] = df_astro['diskon'].fillna('0%').str.replace('%', '').astype(float)

df_astro.head(5)
```
